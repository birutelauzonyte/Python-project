Using the `pandas` library, read the given data file and display the data.

```python
# Kodą rašykite čia...
import pandas as pd
df = pd.read_csv('Birutė.csv', encoding='utf-8')
df
```

Add a new column `Vidurkis` and calculate the average for each student and round it to one decimal place.

```python
# Kodą rašykite čia...

df['Vidurkis'] = (df['Matematika'] + df['Fizika'] + df['Biologija'] + df['Chemija'] +
                  df['Istorija'] + df['Geografija'] + df['Lietuvių kalba'] +
                  df['Anglų kalba'] + df['Informatika'] + df['Ekonomika'] +
                  df['Kūno kultūra'] + df['Muzika'] + df['Dailė'] + df['Technologijos']) / 14

df['Vidurkis'] = df['Vidurkis'].round(1)
df

```

Find and print the name and surname of the best student and his/her average.

```python
# Kodą rašykite čia...


geriausias_studentas = df.loc[df['Vidurkis'].idxmax()]

vardas = geriausias_studentas['Vardas']
pavarde = geriausias_studentas['Pavardė']
vidurkis = geriausias_studentas['Vidurkis']

print(f"The best student is {vardas} {pavarde} with an average of {vidurkis:.2f}.")

       
```

Create a function `stipendija_pagal_vidurki` that would accept a student's average (as an argument) and return the amount of the scholarship awarded to him  based on his average.
* When the average is greater than or equal to 9.5 scholarship 200
* When the average is greater than or equal to 9.0, but less than 9.5 scholarship 180
* When the average is greater than or equal to 8.5, but less than 9.0 scholarship 100
* If the average is less than 8.5, the student does not receive a scholarship (0)

```python
def stipendija_pagal_vidurki(vidurkis):
    
    if vidurkis >= 9.5:
        return 200
    elif vidurkis >= 9.0:
        return 180
    elif vidurkis >= 8.5:
        return 100
    else:
        return 0

print(stipendija_pagal_vidurki)
```

Add a new column `Stipendija` and fill the column with data using the `stipendija_pagal_vidurki` function based on the student's average.

```python
# Kodą rašykite čia...

df['Stipendija'] = df['Vidurkis'].apply(stipendija_pagal_vidurki)
df
```

Count how many students receive which scholarship and print the result.

```python
# Kodą rašykite čia...
stipendija_count = df.groupby('Stipendija').size()
print(stipendija_count)
```
